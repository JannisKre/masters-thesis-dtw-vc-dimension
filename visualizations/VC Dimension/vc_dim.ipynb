```python
import matplotlib.pyplot as plt

def draw_real_axis(ax, xlim=(-2, 2)):
    ax.hlines(0, xlim[0], xlim[1], color="black", linewidth=1.5)
    ax.set_xlim(*xlim)
    ax.set_ylim(-0.45, 0.55)
    ax.set_yticks([])
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)

def draw_points(ax, xs, labels):
    for x, lab in zip(xs, labels):
        color = "tab:green" if lab == "+" else "tab:red"
        ax.scatter(x, 0, s=90, color=color, zorder=3)
        ax.text(x, 0.12, lab, ha="center", va="bottom", fontsize=13, weight="bold")

def draw_interval(ax, left, right, y=-0.0, height=0.1, serif=0.05, color="tab:blue", linewidth=2.8, label=None):
    ax.hlines(y, left, right, color=color, linewidth=linewidth, zorder=2, label=label)
    ax.vlines([left, right], y - height / 2, y + height / 2, color=color, linewidth=linewidth, zorder=2)
    ax.hlines([y - height / 2, y + height / 2], left, left + serif, color=color, linewidth=linewidth, zorder=2)
    ax.hlines([y - height / 2, y + height / 2], right - serif, right, color=color, linewidth=linewidth, zorder=2)

fig, axes = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# 1) Two points: -1 labeled -, +1 labeled +, interval [0.5, 1.5]
ax = axes[0]
draw_real_axis(ax)
draw_interval(ax, 0.5, 1.5, label="interval [0.5, 1.5]")
draw_points(ax, [-1, 1], ["-", "+"] )
ax.set_title("Intervals on ℝ: shattering 2 points (example 1)")
ax.legend(loc="upper right", frameon=False)

# 2) Same two points, both labeled +, interval [-1.2, 1.2]
ax = axes[1]
draw_real_axis(ax)
draw_interval(ax, -1.2, 1.2, label="interval [-1.2, 1.2]")
draw_points(ax, [-1, 1], ["+", "+"] )
ax.set_title("Shattering 2 points (example 2)")
ax.legend(loc="upper right", frameon=False)

# 3) Three points: -1:+, 0:-, 1:+ (impossible for a single interval)
ax = axes[2]
draw_real_axis(ax)
draw_points(ax, [-1, 0, 1], ["+", "-", "+"] )
ax.text(0, -0.25, "Impossible with one interval on ℝ", color="crimson",
        ha="center", fontsize=11, weight="bold")
ax.set_title("3-point labeling not realizable by one interval")

axes[2].set_xticks([-1.2, -1, 0, 1, 1.2])
axes[2].set_xlabel("Real line")
plt.tight_layout()
plt.show()
```
