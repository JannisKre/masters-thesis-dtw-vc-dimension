```python
import matplotlib.pyplot as plt

def plot_s_max(data):
    """
    Plot s_max versus k from input like [[1, n1], [2, n2], ...].

    Parameters
    ----------
    data : list[list[float, float]] | list[tuple[float, float]]
        Sequence of [k, s_max] pairs.

    Returns
    -------
    fig, ax : matplotlib figure and axes
    """
    if not data:
        raise ValueError("Input data must be a non-empty list of [k, s_max] pairs.")

    k_vals, s_vals = zip(*data)

    plt.style.use("seaborn-v0_8-whitegrid")
    fig, ax = plt.subplots(figsize=(7, 4.5), dpi=140)

    ax.plot(
        k_vals,
        s_vals,
        color="#1f77b4",
        linewidth=2.2,
        marker="o",
        markersize=5.5,
        markerfacecolor="white",
        markeredgewidth=1.5,
    )
    ax.set_ylim(bottom=2)
    ax.set_ylim(top=max(s_vals) + 2)
    
    ax.set_xlabel(r"$m$", fontsize=13)
    ax.set_ylabel(r"$s_{\max}$", fontsize=13)
    ax.tick_params(labelsize=11)
    ax.margins(x=0.03, y=0.08)

    fig.tight_layout()
    plt.show()

    return fig, ax


def plot_pi_hat_vs_two_power(data):
    """
    Compare empirical growth data Pi_hat(s) to 2^s.

    Parameters
    ----------
    data : list[list[float, float]] | list[tuple[float, float]]
        Sequence of [s, Pi_hat(s)] pairs.

    Returns
    -------
    fig, ax : matplotlib figure and axes
    """
    if not data:
        raise ValueError("Input data must be a non-empty list of [s, Pi_hat(s)] pairs.")

    s_vals, pi_hat_vals = zip(*data)
    two_pow_vals = [2 ** s for s in s_vals]

    plt.style.use("seaborn-v0_8-whitegrid")
    fig, ax = plt.subplots(figsize=(7.5, 4.8), dpi=140)

    ax.plot(
        s_vals,
        pi_hat_vals,
        label=r"$\Pi(s)$",
        color="#1f77b4",
        linewidth=2.4,
        marker="o",
        markersize=6,
        markerfacecolor="white",
        markeredgewidth=1.5,
    )

    ax.plot(
        s_vals,
        two_pow_vals,
        label=r"$2^s$",
        color="#d62728",
        linewidth=2.0,
        linestyle="--",
        marker="s",
        markersize=5.5,
        markerfacecolor="white",
        markeredgewidth=1.3,
    )

    ax.set_xlabel(r"$s$", fontsize=13)
    ax.set_ylabel("Value", fontsize=13)
    ax.tick_params(labelsize=11)
    ax.legend(fontsize=11)
    ax.margins(x=0.03, y=0.08)

    fig.tight_layout()
    plt.show()

    return fig, ax
```

```python
# From your run log: Pi_hat(s) = 2, 4, 8, 16, 29, 59, 97, 158 for s = 1..8
pi_hat_data = [
    [1, 2],
    [2, 4],
    [3, 8],
    [4, 16],
    [5, 29],
    [6, 59],
    [7, 97],
    [8, 158],
    [9, 233]
]

plot_pi_hat_vs_two_power(pi_hat_data)
```
